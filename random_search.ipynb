```python
from random import choice
from darwin.charles import Population
from darwin.sudoku import Sudoku
from darwin.solve_sudoku import solve_sudoku
from sudoku_templates.templates_final import easy, normal, hard, evil, ultimate

population_sizes = [50, 100, 150, 200, 250, 500, 750, 1000, 1500, 2000, 5000]
templates = [easy[1],  normal[1],  hard[1], evil[1], ultimate[1]]
selection_methods = ['tournament', 'fps', 'boltzmann']
tournament_sizes = [0.01, 0.03, 0.05, 0.07, 0.09, 0.11, 0.13, 0.15, 0.17, 0.19] #percentage
Ts =[0.1, 1, 2 , 5, 10, 100]
cooling_rates = [1.01, 0.99, 0.98, 0.9]
elitisms = [True,False]
elitism_percentages = [0.01, 0.03, 0.05, 0.07, 0.09, 0.11, 0.13, 0.15, 0.17, 0.19]
crossover_rates = [0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]
crossover_methods = ['mixed_single_point_crossover','mixed_multi_point_crossover', 'uniform_crossover', 'a_bit_of_all_crossover']
mutation_rates = [0.01, 0.03, 0.05, 0.07, 0.09, 0.11, 0.13, 0.15, 0.17, 0.19]
mutation_methods = ['mixed_exchange_mutation', 'flip_mutation', 'box_exchange_mutation','a_bit_of_all_mutation']
```

```python
for i in range(1, 200):

    size = choice(population_sizes)
    selection_method = choice(selection_methods)
    
    
    if selection_method == 'fps':
        elitism = True
        elitism_percentage = choice(elitism_percentages)
        tournament_size = None
        T=None
        cooling_rate = None
        
        
    elif selection_method == 'boltzmann':
        T = choice(Ts)
        tournament_size = None
        cooling_rate = choice(cooling_rates)
        elitism = choice(elitisms)
        if elitism:
            elitism_percentage = choice(elitism_percentages)
        else:
            elitism_percentage = None
        
    
    elif selection_method == 'tournament':
        elitism = choice(elitisms)
        if elitism == True:
            elitism_percentage = choice(elitism_percentages)
        else:
            elitism_percentage = None
        tournament_size = choice(tournament_sizes)
        T = None
        cooling_rate = None
        
    crossover_rate = choice(crossover_rates)
    crossover_method = choice(crossover_methods)
    mutation_rate = choice(mutation_rates)
    mutation_method = choice(mutation_methods)
    generations_earlystop = 150
    
    for j in range(len(templates)):

        template = templates[j]
        if j ==0:
            template_diff = 'easy'
        
        elif j ==1:
            template_diff = 'normal'
        
        elif j ==2:
            template_diff = 'hard'
        
        elif j ==3:
            template_diff = 'evil'
        
        elif j ==4:
            template_diff = 'ultimate'
        
        
        else:
            raise Typerror('Something went wrong!')
        
        p = solve_sudoku(population_size= size,
            sudoku = Sudoku(template=template), 
            generations = 500, 
            selection_method =selection_method,
            tournament_size = tournament_size,
            T = T,
            cooling_rate= cooling_rate,
            elitism = elitism,
            elitism_percentage = elitism_percentage,
            crossover_rate = crossover_rate,
            crossover_method = crossover_method,
            mutation_rate = mutation_rate,
            mutation_method = mutation_method, 
            log=True, 
            log_name= f"random_search_final/{i}-{template_diff}-{size}-{selection_method}-{elitism}-{elitism_percentage}-{tournament_size}-{T}-{cooling_rate}-{crossover_rate}-{crossover_method}-{mutation_rate}-{mutation_method}-{generations_earlystop}",
            generations_earlystop=generations_earlystop
        )
```
